### 统计不同样本的个数
- 将测试集和训练集中每个 label 的样本数相加
- 将结果绘制为柱状图

```python
import numpy as np
```

```python
# 定义公共变量
label_file_train = "../npy_data/multi/train/label.npy" # 训练集的 label
label_file_test = "../npy_data/multi/test/label.npy" # 训练集的 label

label2index = {'BitTorrent': 0, 'Cridex': 1, 'Facetime': 2, 'FTP': 3, 'Geodo': 4, 'Gmail': 5, 'Hthot': 6, 'Miuref': 7, 'MySQL': 8, 'Neris': 9, 'Nsis': 10, 'Outlook': 11, 'Shifu': 12, 'Skype': 13, 'SMB': 14, 'Tinba': 15, 'Virut': 16, 'Weibo': 17, 'WorldOfWarcraft': 18, 'Zeus': 19}

index2label = {0: 'BitTorrent', 1: 'Cridex', 2: 'Facetime', 3: 'FTP', 4: 'Geodo', 5: 'Gmail', 6: 'Hthot', 7: 'Miuref', 8: 'MySQL', 9: 'Neris', 10: 'Nsis', 11: 'Outlook', 12: 'Shifu', 13: 'Skype', 14: 'SMB', 15: 'Tinba', 16: 'Virut', 17: 'Weibo', 18: 'WorldOfWarcraft', 19: 'Zeus'}
```

```python
label_train_data = np.load(label_file_train) # 获得 train 的 label 文件
label_test_data = np.load(label_file_test) # 获得 test 的 label 文件
```

```python
print('训练集 label 个数 {}; 测试集 label 个数 {};'.format(label_train_data.shape, label_test_data.shape))
```

```python
# 统计 各 类样本的各自的个数
unique, counts = np.unique(label_test_data, return_counts=True)
test_label2num = dict(zip(unique, counts))
print(test_label2num)

unique, counts = np.unique(label_train_data, return_counts=True)
train_label2num = dict(zip(unique, counts))
print(train_label2num)
```

```python
label2num_train = {} # 每个 label 的样本个数
label2num_test = {}

for label in train_label2num:
    label_name = index2label[label]

    label2num_train[label_name] = train_label2num[label]

for label in test_label2num:
    label_name = index2label[label]
    label2num_test[label_name] = test_label2num[label]

print("train:",label2num_test)
print("test:",label2num_test)
```

```python
label2num_train = {} # 每个 label 的样本个数
label2num_test = {}

for label in train_label2num:
    label_name = index2label[label]
    label2num_train[label_name] = train_label2num[label]

for label in test_label2num:
    label_name = index2label[label]
    label2num_test[label_name] = test_label2num[label]

print("train:",label2num_test)
print("test:",label2num_test)
```

```python
label2num_train = {} # 每个 label 的样本个数
label2num_test = {}

for label in train_label2num:
    label_name = index2label[label]
    label2num_train[label_name] = train_label2num[label]

for label in test_label2num:
    label_name = index2label[label]
    label2num_test[label_name] = test_label2num[label]

print("train:",label2num_test)
print("test:",label2num_test)
```

```python
import matplotlib.pyplot as plt
# plt.style.use(['science', 'no-latex'])
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
labels = label2num_train.keys()
data = label2num_train.values()

plt.bar(labels, data) # 绘图
ax.set_xticklabels(labels, rotation=-45, fontsize=12) # 设置字体大小和倾斜

plt.show()
```
